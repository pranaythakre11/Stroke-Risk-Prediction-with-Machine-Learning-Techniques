DATA PROCESSING

```python
#In x axis we are including all the columns except thr targeted rows
x=df.iloc[ : : ,1:11]
x
```

```python
#In y axis we are including only tageted rows
y=df.iloc[ : ,-1]
y
```

ENCODING

```python
#OrdinalENcoder:It is used for encoding categorical features into ordinal integer
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe=OrdinalEncoder()
x[["gender","ever_married","work_type","Residence_type","smoking_status"]]=oe.fit_transform(x[["gender","ever_married","work_type","Residence_type","smoking_status"]])
x
```

MODEL

```python
#train_test_split:It isused to split a dataset into two subsets

from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=1,stratify=y)
```

```python
#LogesticRegression:It is used to perform Logestic Regression
from sklearn.linear_model import LogisticRegression
```

```python
#classification_report:It provide the comprehensive report with metrics such as precision ,recall,F1-score.
#confusion_matrix:It provides a matrix that compares the predicted class labels against the actual class labels.
from sklearn.metrics import classification_report,confusion_matrix
```

```python
#step 1:Import the model
from sklearn.linear_model import LinearRegression

#step 2:initalize the model
linreg=LinearRegression()

#step 3:train the model
linreg.fit(xtrain,ytrain)

#step 4:make prediction
ypred=linreg.predict(xtest)
```

```python
#Accuracy:It is used to see how accurate the data is provided
from sklearn.metrics import r2_score
print(f"Accuracy : {r2_score(ytest,ypred)}")
```

```python
#linreg.intercept_:it provide the value of the intercept
linreg.intercept_
```

```python
#It represent the coefficient associated with each input feature in the linear regression equation
linreg.coef_
```
